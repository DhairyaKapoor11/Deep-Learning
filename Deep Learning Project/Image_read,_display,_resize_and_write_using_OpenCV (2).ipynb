```
# Horizontal and Vertical Flipping of images
```

```
import cv2
```

```
image = cv2.imread('apple.jpg')
cv2_imshow(image)
```

```
hflip = cv2.flip(image, 1)
cv2_imshow(hflip)
```

```
hflip = cv2.flip(image, 0)
cv2_imshow(hflip)
```

```
hflip = cv2.flip(image, -1)
cv2_imshow(hflip)
```
